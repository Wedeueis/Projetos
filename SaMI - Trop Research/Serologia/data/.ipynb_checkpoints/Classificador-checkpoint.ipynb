```python
cols = ['Plate&Well', 'IBAG35r',
       'IBAG36r', 'IBAG37r', 'IBAG38r', 'IBAG39r', 'IBAG95r', 'IBAG97r',
       'IBAG99r', 'IBAG101r', 'IBAG112r', 'IBAG20r', 'IBAG127r', 'IBAG110r',
       'IBAG108r', 'IBAG131r']
df = df[cols].values
```

```python
labels = {}
antigen = [0, 6, 12, 18, 24, 30, 1, 7, 13, 19, 25, 31, 2, 8, 14, 20, 26, 32, 3, 9, 15, 21, 27, 33, 4, 10, 16, 22, 28, 34]

for i in range(len(df)):
    for j in range(len(df[i])-1):
        try:
            #valor atribuido (media dos dois spots)
            value = float(df[i][j+1])

            #spot 1
            idx = 2*j
            name = df[i][0] + "_antigen" + str(antigen[idx]) + ".png"
            print(name)
            labels[name] = value
            #spot 2
            idx = idx+1
            name = df[i][0] + "_antigen" + str(antigen[idx]) + ".png"
            print(name)
            labels[name] = value
        except:
            print("error creating sample pair for image: {}".format(df[i][0]))
            
```

```python
import os
import cv2
```

```python
path = '/media/wedeueis/DATA/data/antigen_spots/'

X = []
y = []

for filename in os.listdir(path):
    file = path + filename
    img = cv2.imread(file, 0)
    img = cv2.resize(img, (50,50), interpolation= cv2.INTER_AREA)
    if img is not None:
        print("Processing file: " + filename)
        print("label: " + str(labels[filename]) )
        X += [img]
        y += [labels[filename]]
        if labels[filename] == 0:
            cv2.imwrite("spots/00/"+filename[:-4]+".jpg", img)
        elif labels[filename] == 0.5:
            cv2.imwrite("spots/05/"+filename[:-4]+".jpg", img)
        elif labels[filename] == 1:
            cv2.imwrite("spots/10/"+filename[:-4]+".jpg", img)
        elif labels[filename] == 2:
            cv2.imwrite("spots/20/"+filename[:-4]+".jpg", img)
        elif labels[filename] == 3:
            cv2.imwrite("spots/30/"+filename[:-4]+".jpg", img)
        else:
            print("error saving image")
```

```python
#### Test ######
print("Quantidade de imagens: " + str(len(X)) )
print("Quantidade de labels: " + str(len(y)) )
plt.imshow(X[0], cmap="gray")
print("Label: " + str(y[0]) )
```

```python
from sklearn.model_selection import train_test_split
from sklearn import svm, metrics
from sklearn.preprocessing import LabelEncoder
```

```python
X = np.array(X)
y = np.array(y)

n_samples = len(X)
X_reshaped = X.reshape((n_samples, -1))

le = LabelEncoder()

y_encoded = le.fit_transform(y)

print(list(le.classes_))

X_train, X_test, y_train, y_test = train_test_split(X_reshaped, y_encoded, test_size=0.3, random_state=42)
```

```python
classifier = svm.SVC(gamma=0.001)

classifier.fit(X_train, y_train)
```

```python
predicted = classifier.predict(X_test)
```

```python
print("Classification report for classifier %s:\n%s\n"
      % (classifier, metrics.classification_report(y_test, predicted)))
print("Confusion matrix:\n%s" % metrics.confusion_matrix(y_test, predicted))
```

```python
from sklearn.neural_network import MLPClassifier

clf = MLPClassifier(solver='lbfgs', alpha=1e-5,
                    hidden_layer_sizes=(100, 5), random_state=1, activation='relu', max_iter=100000)
```

```python
print(clf.fit(X_train, y_train))    
```

```python
result = clf.predict(X_test)
```

```python
print("Classification report for classifier %s:\n%s\n"
      % (clf, metrics.classification_report(y_test, result)))
print("Confusion matrix:\n%s" % metrics.confusion_matrix(y_test, result))
```
