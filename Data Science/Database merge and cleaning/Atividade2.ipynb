```python
import numpy as np
import pandas as pd
```

#### Leitura dos dados

```python
ibge2010 = pd.read_csv("ibge2010.csv")
ibge2011 = pd.read_csv("ibge2011.csv")
ibge2012 = pd.read_csv("ibge2012.csv")
mesoregioes = pd.read_csv("mesorregioes.csv")
```

#### Encontrar e preencher posições vazias

```python
np.where(pd.isnull(ibge2010))
```

```python
np.where(pd.isnull(ibge2011))
```

```python
np.where(pd.isnull(ibge2012))
```

```python
# Calculo e preenchimento do PIB per capta para posições vazias
# como demonstrado nas células anterioes apenas a tabela de 2011 apresenta posições vazias
# na coluna pib per capta (coluna 6)
for i,j in zip(*np.where(pd.isnull(ibge2011))):
    pib = ibge2011["PIB"][i]
    pop = ibge2011["Populacao"][i]
    mun = ibge2011["NomeMunicipio"][i]
    pippercapta = pib / pop
    print("Idx.: {} {} Município: {}  PIB: {:.2f} Pop.: {} PIB/capta: {:.2f}".format (i, j, mun, pib, pop, pippercapta) )
    ibge2011.iloc[i,j] = pippercapta
```

#### Fundir as colunas faltantes

```python
# A tabela de 2012 não possui a coluna de Pib per capta
ibge2012.columns
```

```python
# Podemos calcular os valores usando as colunas PIB e População, acrescentando uma nova coluna
ibge2012["PIBperCapita"] = ibge2012["PIB"] / ibge2012["Populacao"]
```

```python
print(ibge2012["PIBperCapita"])
```

```python
# A tabela de 2011 não possui a coluna NomeMeso
ibge2011.columns
```

```python
# Podemos usar a tabela mesoregioes com o valor de NomeMunicipio como chave para mapear os valores
mesoregioes.columns
```

```python
ibge2011 = pd.merge(ibge2011, mesoregioes)
```

```python
ibge2011.columns
```

```python
print(ibge2011)
```

#### Consolidando os dados em um único arquivo

```python
ibge_consolidado = pd.concat([ibge2010, ibge2011, ibge2012])
```

```python
ibge_consolidado
```

```python
ibge_consolidado.to_csv("./ibge_consolidado.csv")
```

#### Dados por Mesoregião

```python
ibge_meso = ibge_consolidado.groupby(["Ano", "NomeMeso"])["PIB", "Populacao"].sum()
```

```python
ibge_meso
```

```python
ibge_meso["PIBperCapita"] = ibge_meso["PIB"] / ibge_meso["Populacao"]
```
