1. O arquivo “NoisyGull.pgm” contém uma imagem corrompida por ruídos. Desenvolva um algoritmo  que permita aplicar o filtro da média com kernel de tamanhos 2×2, 5×5, 10×10 e 20×20. Mostre o resultado de cada kernel aplicado e quantifique os resultados através do erro médio quadrático. 

```python
# sua solução
import numpy as np
import cv2
import matplotlib.pyplot as plt
%matplotlib inline

def spatial_filter(img, kernel):
    temp = img.copy()
    l, c = img.shape
    lm, cm = kernel.shape
    normalizer = np.sum(kernel)
    for i in range(l):
        for j in range(c):
            value = 0
            for s in range(lm):
                for t in range(cm):
                    x = int(i+s-lm/2+.5)
                    y = int(j+t-cm/2+.5)
                    if x>=0 and x<l and y>=0 and y<c:
                        value += img[x,y]*kernel[s,t]
            value = value/normalizer
            temp[i,j] = value
    return temp

def mse(image1, image2):
    err = np.sum((image1.astype("float") - image2.astype("float")) ** 2)
    err /= float(image1.shape[0] * image1.shape[1])
    return err

def showfig(img, ucmap='gray'):
    plt.figure(figsize=(5,5))
    imgplot=plt.imshow(img, ucmap)
```

```python
img = plt.imread("data/NoisyGull.pgm")
print(img.shape)
showfig(img)
kernel1 = np.ones((2,2))
denoised = spatial_filter(img, kernel1)
showfig(denoised)
erro = mse(img, denoised)
print("Erro do filtro com kernel 2x2: ", erro)

kernel2 = np.ones((5,5))
denoised = spatial_filter(img, kernel2)
showfig(denoised)
erro = mse(img, denoised)
print("Erro do filtro com kernel 5x5: ", erro)

kernel3 = np.ones((10,10))
denoised = spatial_filter(img, kernel3)
showfig(denoised)
erro = mse(img, denoised)
print("Erro do filtro com kernel 10x10: ", erro)

kernel4 = np.ones((20,20))
denoised = spatial_filter(img, kernel4)
showfig(denoised)
erro = mse(img, denoised)
print("Erro do filtro com kernel 20x20: ", erro)
```

3. Calcule a filtragem espacial para a seguinte imagem:

Imagem
        
0	1	2	3	5

1	2	3	5	7

2	3	5	7	8

3	5	7	8	9

5	7	8	9	10

Kernel
        
1	1	1

1	2	1

1	1	1


```python
# sua solução
def spatial_filter2(image, mask):
    ilin, icol = image.shape
    mlin, mcol = mask.shape
    n = mlin*mcol
    mlin = int(mlin/2)
    mcol = int(mcol/2)
    newimage = image.copy()
    for i in range(0,ilin):
        for j in range(0,icol):
            temp = 0
            for k in range (-mlin, mlin):
                for l in range(-mcol,mcol):
                    vi = i+k
                    vj = j+l
                    if( (vi >= 0) & (vi < ilin) & (vj>=0) & (vj<icol) ):
                        temp = temp + image[vi][vj]*mask[k+mlin][l+mcol]
            
            newimage[i][j] = temp
    return newimage

img = np.array([[0,1,2,3,5],
                [1,2,3,5,7],
                [2,3,5,7,8],
                [3,5,7,8,9],
                [5,7,8,9,10]], dtype=np.float)
kernel = np.array([[1,1,1],
                   [1,2,1],
                   [1,1,1]], dtype=np.float)
                
processed = spatial_filter2(img, kernel)
showfig(img)
showfig(processed)
print(processed)
```

5. Implemente um algoritmo para equalizar uma imagem retornando-a equalizada. Aplique sobre um conjunto de imagens. NÃO UTILIZE FUNÇÕES PRONTAS.

```python
# sua solução
def histogram(image):
    hist = [0]*256
    lin,col = image.shape
    for i in range(0,lin):
        for j in range(0,col):
            hist[image[i][j]]+=1
    return hist

def cumulativeDist(list):
    n = len(list)
    newlist = [0]*n
    newlist[0] = list[0]
    for i in range(1,n):
        newlist[i] = newlist[i-1] + list[i]
    return newlist

def normalize(list, k):
    n = len(list)
    newlist = [0]*n
    for i in range(0,n):
        newlist[i] = list[i]/k
    return newlist

def equalizeHist(image):
    hist = histogram(image)
    lin,col = image.shape
    p = normalize(hist,lin*col)
    s = cumulativeDist(p)
    table = [0]*256
    for k in range(0,len(table)):
        table[k] = int(round(s[k]*255))
    newimage = image.copy()
    for i in range(0,lin):
        for j in range(0,col):
            newimage[i][j] = table[ image[i][j] ]
    return newimage

img = cv2.imread("data/house.tif",0)
img2 = cv2.imread("data/lc_einstein.tif",0)
img3 = cv2.imread("data/woman.tif",0)

eq_img = equalizeHist(img)
eq_img2 = equalizeHist(img2)
eq_img3 = equalizeHist(img3)

showfig(img,"gray")
showfig(eq_img,"gray")
showfig(img2,"gray")
showfig(eq_img2,"gray")
showfig(img3,"gray")
showfig(eq_img3,"gray")
```
